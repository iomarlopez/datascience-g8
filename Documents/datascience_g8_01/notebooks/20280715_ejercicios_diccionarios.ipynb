```python
# Ejercicio 1

inventario_norte = {"laptop": 4, "mouse": 10, "monitor": 3} 
inventario_sur = {"laptop": 2, "teclado": 8, "monitor": 5} 
inventario_centro = {"mouse": 6, "teclado": 4, "webcam": 7} 
```

```python
inventario_total = {}

for llave, valor in inventario_norte.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor


for llave, valor in inventario_sur.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor


for llave, valor in inventario_centro.items():
    if llave not in inventario_total:
        inventario_total[llave] = 0
    inventario_total[llave] += valor
```

```python
print(inventario_total)
```

```python
llaves_ordenadas = sorted(inventario_total.keys())
inventario_total_con_llaves_ordenadas = {}

for llave in llaves_ordenadas:
    inventario_total_con_llaves_ordenadas[llave] = inventario_total[llave]

print(inventario_total)
print(inventario_total_con_llaves_ordenadas)
```

```python
#Ejercicio 2
calificaciones = { 
"Ana": [17, 18, 16, 19], 
"Luis": [15, 16, 17, 16], 
"Carla": [18, 17, 19, 18], 
"Pedro": [16, 16, 16, 16], 
"Lucía": [15, 17, 16, 16] 
}

estadisticas = {}

for estudiante, notas in calificaciones. items():
    print(estudiante, notas)
    estadisticas[estudiante] = {
        "minima_nota": min(notas),
        "maxima_nota": max(notas),
        "promedio": sum(notas)/len(notas)
    }
    print("="*20)
```

```python
calificaciones.items()
```

```python
for estudiante, stats in estadisticas.items():
    print(estudiante)
    print(stats)
    print("="*20)
```

```python
estadisticas.items()
```

```python
ranking = []
for nombre, stats in estadisticas.items():
    ranking.append((nombre, stats["promedio"]))

ranking = sorted(ranking, key=lambda x: x[1], reverse=True)
print (ranking)
```

```python
ranking[2:] = sorted(ranking[2:], key=lambda x:x[0])
print(ranking)
```

```python
#Ejercicio 3

usuarios = { 
"  ANA_01 ": {"nombre": " ana torres ", "correo": " ANA@MAIL.COM ", "edad": "24"}, 
"LUIS_02": {"nombre": "luis perez", "correo": "Luis@Mail.com", "edad": "treinta"}, 
" carla_03": {"nombre": "CARLA RUIZ", "correo": "carla@mail.com", "edad": "27"} 
}

usuarios_limpios = {}

for usuario, datos in usuarios.items():
    print(usuario, datos)

    llave_limpia = usuario.strip()
    correo_minuscula = datos["correo"].lower()
    nombre_limpio = datos["nombre"].strip().upper()
    
    try:
        edad_entera = int(datos["edad"])
    except:
        edad_entera = None

    datos_limpios = {
        "nombre": nombre_limpio,
        "correo": correo_minuscula,
        "edad": edad_entera
    }

    usuarios_limpios[llave_limpia] = datos_limpios

    print("="*20)
```

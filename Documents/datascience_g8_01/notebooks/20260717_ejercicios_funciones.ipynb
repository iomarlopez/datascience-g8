```python
def analizar_ventas(ventas: list[int | float])-> dict[str, int | float]:
    if not ventas:
        return "No hay ventas"
    
    for venta in ventas:
        if venta < 0:
            return f"Hay ventas negativas: {ventas}"
        
    total_vendido = sum(ventas)
    promedio = total_vendido / len(ventas)
    venta_mas_alta = max(ventas)
    venta_mas_baja = min(ventas)

    total_superior_promedio = 0
    total_ventas_inferior_100=0

    contar = 0
    for venta in ventas:
        if venta > promedio:
            total_superior_promedio = total_superior_promedio +1

        if venta < 100:
            total_ventas_inferior_100 = total_ventas_inferior_100 + 1
    
    return{
        "total_vendido": total_vendido,
        "promedio": promedio,
        "venta_mas_alta": venta_mas_alta,
        "venta_mas_baja": venta_mas_baja,
        "total de ventas superior al promedio": total_superior_promedio,
        "total de ventas inferior a 100": total_ventas_inferior_100
    }

analizar_ventas(ventas)
```

```python
#Ejercicio 3

def registrar_estudiante(estudiantes, nombre, notas):
    # Validar nombre
    if not nombre.strip():
        return "Error: El nombre no puede estar vacío."
    if nombre in estudiantes:
        return f"Error: El estudiante '{nombre}' ya está registrado."

    # Validar cantidad de notas
    if len(notas) < 3:
        return "Error: El estudiante debe tener al menos tres notas."

    # Validar notas numéricas entre 0 y 20
    for nota in notas:
        if not isinstance(nota, (int, float)):
            return "Error: Todas las notas deben ser numéricas."
        if nota < 0 or nota > 20:
            return "Error: Las notas deben estar entre 0 y 20."

    # Calcular promedio con dos decimales
    promedio = round(sum(notas) / len(notas), 2)

    # Determinar estado
    estado = "Aprobado" if promedio >= 11 else "Desaprobado"

    # Guardar en el diccionario
    estudiantes[nombre] = {
        "notas": notas,
        "promedio": promedio,
        "estado": estado
    }

    return f"Registro exitoso: {nombre} ({estado}, Promedio: {promedio})"


# Datos de prueba
estudiantes = {}
print(registrar_estudiante(estudiantes, 'Ana', [18, 15, 17]))
print(estudiantes)
```

```python
# Ejercicio 4

def agrupar_por_longitud(palabras):
    resultado = {}
    for palabra in palabras:
        if not isinstance(palabra, str):
            continue
        palabra_normalizada = palabra.lower()
        longitud = len(palabra_normalizada)
        if longitud not in resultado:
            resultado[longitud] = set()
        resultado[longitud].add(palabra_normalizada)
    for longitud in resultado:
        resultado[longitud] = sorted(resultado[longitud])
    return resultado

# Datos de prueba
palabras = ['python', 'Java', 'código', 'web', 'java', 'datos', 'API', 'lista']
resultado = agrupar_por_longitud(palabras)
print(resultado)
```

```python
# Ejercicio 5

def actualizar_inventario(inventario, producto, cantidad, operacion):
    if operacion not in ["entrada", "salida"]:
        return {"mensaje": "Error: Operación inválida.", "stock_actual": inventario.get(producto, 0)}

    if not isinstance(cantidad, int) or cantidad <= 0:
        return {"mensaje": "Error: La cantidad debe ser un entero positivo.", "stock_actual": inventario.get(producto, 0)}

    if operacion == "entrada":
        inventario[producto] = inventario.get(producto, 0) + cantidad
        return {"mensaje": f"Entrada registrada: {cantidad} {producto}(s).", "stock_actual": inventario[producto]}

    if operacion == "salida":
        if producto not in inventario:
            return {"mensaje": "Error: Producto no existe.", "stock_actual": 0}
        if inventario[producto] < cantidad:
            return {"mensaje": "Error: Stock insuficiente.", "stock_actual": inventario[producto]}
        inventario[producto] -= cantidad
        return {"mensaje": f"Salida registrada: {cantidad} {producto}(s).", "stock_actual": inventario[producto]}


#Datos de prueba
inventario = { 'Laptop': 5, 'Mouse': 12, 'Teclado': 8 }
resultado = actualizar_inventario(inventario, 'Mouse', 4, 'salida')
print(resultado)
print(inventario)
```

```python
#Ejercicio 6

def calcular_pedido(productos, cantidades, descuento=0, delivery=0):
    resultado = {
        "subtotal": 0,
        "descuento": 0,
        "delivery": 0,
        "total": 0,
        "no_encontrados": []
    }

    if not isinstance(descuento, (int, float)) or descuento < 0 or descuento > 100:
        descuento = 0

    for producto, cantidad in cantidades.items():
        if not isinstance(cantidad, int) or cantidad <= 0:
            resultado["no_encontrados"].append(producto)
            continue

        if producto not in productos:
            resultado["no_encontrados"].append(producto)
            continue

        precio = productos[producto]
        resultado["subtotal"] += precio * cantidad

    resultado["descuento"] = round(resultado["subtotal"] * (descuento / 100), 2)
    resultado["delivery"] = delivery
    resultado["total"] = round(resultado["subtotal"] - resultado["descuento"] + resultado["delivery"], 2)

    return resultado

#Datos de prueba
productos = {'Hamburguesa': 15, 'Pizza': 25, 'Gaseosa': 5}
cantidades = {'Hamburguesa': 2, 'Gaseosa': 3, 'Helado': 1}

resultado = calcular_pedido(productos, cantidades, descuento=10, delivery=5)
print(resultado)
```

```python
# Ejercicio 7

def evaluar_empleados(empleados, minimo=70):
    resultado = []
    for nombre, notas in empleados.items():
        if not notas:
            continue
        promedio = round(sum(notas) / len(notas), 2)

        if promedio >= 90:
            clasificacion = "Excelente"
        elif promedio >= 80:
            clasificacion = "Bueno"
        elif promedio >= 70:
            clasificacion = "Regular"
        else:
            clasificacion = "Deficiente"

        if promedio >= minimo:
            resultado.append((nombre, promedio, clasificacion))

    resultado.sort(key=lambda x: (-x[1], x[0]))
    return resultado


#Datos de prueba

empleados = {
    'Ana': [80, 90, 75],
    'Carlos': [60, 70, 65],
    'Lucía': [95, 85, 90],
    'Pedro': [70, 70, 70]
}

resultado = evaluar_empleados(empleados)
print(resultado)
```

```python
#Ejercicio 8

def comprimir_texto(texto):
    if not texto:
        return ""

    resultado = []
    contador = 1

    for i in range(1, len(texto)):
        if texto[i] == texto[i-1]:
            contador += 1
        else:
            resultado.append(texto[i-1] + str(contador))
            contador = 1

    # último grupo
    resultado.append(texto[-1] + str(contador))

    return "".join(resultado)


def descomprimir_texto(texto):
    if not texto:
        return ""

    resultado = []
    i = 0
    while i < len(texto):
        if not texto[i].isalpha():
            return "Error: Formato inválido."
        letra = texto[i]
        i += 1

        if i >= len(texto) or not texto[i].isdigit():
            return "Error: Formato inválido."

        numero = ""
        while i < len(texto) and texto[i].isdigit():
            numero += texto[i]
            i += 1

        cantidad = int(numero)
        resultado.append(letra * cantidad)

    return "".join(resultado)


# Datos de prueba
texto = 'aaabbccccdaa'
comprimido = comprimir_texto(texto)
original = descomprimir_texto(comprimido)

print("Texto original:", texto)
print("Texto comprimido:", comprimido)
print("Texto descomprimido:", original)

```

```python
#Ejercicio 10

def reservar_asientos(sala, asientos_solicitados):
    sala_temp = [fila[:] for fila in sala]

    for fila, columna in asientos_solicitados:
        if fila < 1 or fila > len(sala) or columna < 1 or columna > len(sala[0]):
            return {"estado": "Error: Asiento fuera de rango.", "reservados": [], "sala": sala}

        f, c = fila - 1, columna - 1

        if sala[f][c] != 'L':
            return {"estado": "Error: Asiento ocupado.", "reservados": [], "sala": sala}

        sala_temp[f][c] = 'O'

    for fila in range(len(sala)):
        sala[fila] = sala_temp[fila]

    return {"estado": "Reserva exitosa", "reservados": asientos_solicitados, "sala": sala}


def mostrar_sala(sala):
    columnas = len(sala[0])
    encabezado = "   " + " ".join(str(i+1) for i in range(columnas))
    print(encabezado)
    for i, fila in enumerate(sala, start=1):
        print(f"{i:2} " + " ".join(fila))


# Datos de prueba
sala = [['L', 'L', 'O', 'L'],
        ['O', 'L', 'L', 'L'],
        ['L', 'O', 'L', 'L']]

asientos = [(1, 2), (2, 3), (3, 1)]

resultado = reservar_asientos(sala, asientos)
print(resultado["estado"])
print("Reservados:", resultado["reservados"])
mostrar_sala(resultado["sala"])
```

### Funciones u operadores en Strings

- Longitud de mi cadena (len)
- Convertir a mayuscula (upper)
- Convertir a minuscula (lower)
- Reemplazar un valor de la cadena (replace)
- "texto        " -> (strip) -> "texto"
- "texto, texto2, texto3" -> (split, ",") ->[texto1, texto2, texto3]

```python
nombre = "Victor"
print("Cuantas eltras tiene mi nombre: ", len(nombre))

print(len("peru"))
```

```python
print(nombre.upper())
```

```python
print(nombre.lower())
```

```python
##replaxe acepta 2 parametros, primer parametro: lo que quiero reemplzar, segundo parametro: el valor por el que voy a reemplazar 
print(
    nombre.replace("tor", "kram")
)
```

```python
nombre = "Victor         "
print(len(nombre))

nombre_limpio = nombre.strip() ##limpiando resulta en 6 caracteres

print(len(nombre_limpio))

nombre = "       Victor      "
print(len(nombre))
print(len(nombre_limpio)) # 6 caracteres
```

```python
nombres = "Victor Lopez"
nombres_separados = nombres.split(" ")
print(nombres_separados)

nombres = "VictorxxLopez"
nombres_separados = nombres.split("xx")
print(nombres_separados)
```

## Operadores Logicos

- UNION -> | 
- INTERSECCION -> &
- DIFERENCIA -> -
- DIFERENCIA SIMETRICA -> ^

```python
##Ejercicio 1

estudiantes_python = { "Ana", "Luis", "María", "Carlos", "Sofía", "Diego" } 
estudiantes_javascript = { "María", "Carlos", "Pedro", "Lucía", "Diego", "Valeria" }

print("Estudiantes en ambos cursos", estudiantes_python & estudiantes_javascript)
print("Estudiantes en ambos cursos", estudiantes_python.intersection(estudiantes_javascript))

print("Estudiantes solo en python:", estudiantes_python - estudiantes_javascript)

print("Estudiantes solo en javascript:", estudiantes_javascript - estudiantes_python)

print ("Todos los estudiantes:", estudiantes_python | estudiantes_javascript)
```

```python
#Ejercicio 2 
almacen_arequipa = { 
"laptop", "teclado", "mouse", "monitor", "impresora", "audifonos" 
} 
almacen_lima = { 
"laptop", "mouse", "monitor", "tablet", "webcam", "parlantes" 
} 

print("Productos disponibles en ambos almacenes: ", almacen_arequipa & almacen_lima)
print("Productos disponibles solo en Arequipa: ", almacen_arequipa - almacen_lima)
print("Productos disponibles solo en Lima: ", almacen_lima - almacen_arequipa)
print("Todos los productos: ", almacen_arequipa | almacen_lima)

if almacen_lima == almacen_arequipa:
    print("Ambos almacenes tienen el mismo stock"),
else: 
    print("Los almacenes no tienen el mismo stock")
```

```python
# Ejercicio 3
permisos_usuario = { 
"leer", "editar", "comentar", "descargar" 
} 
permisos_requeridos = { 
"leer", "editar", "eliminar", "publicar" 
} 

if permisos_requeridos.issuperset(permisos_usuario):
    print("El usuario cuenta con todos los permisos"),
else: 
    print("El usuario no cuenta con los permisos requeridos")

print("Al usuario le faltan los siguiente permisos: ", permisos_requeridos - permisos_usuario)
print("El usuario tiene los siguiente permisos adicionales: ", permisos_usuario - permisos_requeridos)
print("Se tiene los siguientes permisos diferentes", permisos_requeridos ^ permisos_usuario)
```

```python
## Ejercicio 4
numeros_a = {2, 4, 6, 8, 10, 12, 14} 
numeros_b = {4, 6, 9, 10, 13, 14, 16} 
numeros_c = {1, 4, 6, 10, 11, 14, 18} 


print("Numeros presentes en los 3 conjuntos: ", numeros_a & numeros_b & numeros_c)
print("Los números presentes en numeros_a y numeros_b, pero no en numeros_c son: ", (numeros_a & numeros_b) - numeros_c)
print("Los números que solo pertenecen a numeros_a son: ", numeros_a - (numeros_b | numeros_c))
print("Todos los números diferentes son: ", (numeros_a ^ numeros_b) | (numeros_b ^ numeros_c) | (numeros_c ^ numeros_a)) 
```

```python
# Ejercicio 5
habilidades_candidato_1 = { 
"Python", "Git", "SQL", "Docker", "React" 
} 
habilidades_candidato_2 = { 
"Python", "JavaScript", "React", "Node.js", "Git" 
} 
habilidades_requeridas = { 
"Python", "Git", "SQL", "Docker" 
}



print("Las habilidades que comparten ambos candidatos son :", habilidades_candidato_1 & habilidades_candidato_2)
print("Las habilidades exclusivas de candidato 1 son:", habilidades_candidato_1 - habilidades_candidato_2)
print("Las habilidades exclusivas de candidato 2 son:", habilidades_candidato_2 - habilidades_candidato_1)
print("Las habilidades requeridas que le faltan a candidato 1 son:", habilidades_requeridas - habilidades_candidato_1)
print("Las habilidades requeridas que le faltan a candidato 2 son:", habilidades_requeridas - habilidades_candidato_2)
print ("Todas las habilidades disponibles son:", habilidades_requeridas | habilidades_candidato_1 | habilidades_candidato_2)
```

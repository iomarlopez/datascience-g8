```python
lista_de_letras = ['b', 'a', 'e', 'd']
lista_de_letras.reverse()
print(lista_de_letras)
```

```python
lista_de_letras = ['b', 'a', 'e', 'd']
lista_de_letras_2 = ["c","z","h"]

for letra in lista_de_letras_2:
    # print(letra)
    lista_de_letras.append(letra)

print(lista_de_letras)
```

```python
lista_de_letras = ['b', 'a', 'e', 'd']
lista_de_letras_2 = ["c","z","h"]
lista_de_letras.extend(lista_de_letras_2)
print(lista_de_letras)
```

```python
lista_de_letras = ['b', 'a', 'e', 'd']
lista_de_letras_2 = ["c","z","h"]

nueva_lista = lista_de_letras + lista_de_letras_2
print(nueva_lista)
```

### METODOS DE LISTAS

- insert -< inserta un elemento en la posicion o indice dado
- append -> inserta un elemento en la ultima posicion
- clear -> borra el contenido de mi lista
- pop -> elimina el ultimo elemento de la lista
- extend -> agrega los elelentos de una lista a otra lista
- sort -> ordena los elementos de una lista
- count -> cuenta el valor que se le de en una lista
- len -> sirve para saber cuantos elementos tiene mi lista

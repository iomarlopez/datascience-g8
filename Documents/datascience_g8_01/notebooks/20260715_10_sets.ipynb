```python
# UNION
conjunto_a = {1,3,5,7,9,1,3}
conjunto_b = {3,5,7,12}

conjunto_a.union(conjunto_b)
```

```python
conjunto_a | conjunto_b
```

```python
# INTERSECCION
conjunto_a.intersection(conjunto_b)
```

```python
conjunto_a & conjunto_b
```

```python
# DIFERENCIA
conjunto_a.difference(conjunto_b)
```

```python
conjunto_a - conjunto_b
```

```python
# DIFERENCIA SIMETRICA
conjunto_a.symmetric_difference(conjunto_b)
```

```python
conjunto_a ^ conjunto_b
```

```python
#si tienen algo en común = false
conjunto_a = {1,3,5,7,9,1,3}
conjunto_b = {3,5,7,12}

conjunto_a.isdisjoint(conjunto_b)
```

```python
#si no tienen nada en comun = true
conjunto_a = {1,3,5,7,9,1,3}
conjunto_b = {180}

conjunto_a.isdisjoint(conjunto_b)
```

```python
mi_diccionario = { 
    "peru": "lima",
    "argentina": "buenos aires",
    "noruega": "oslo",
    "japon": "tokio"
}

print(len(mi_diccionario))
```

```python
for key in mi_diccionario:
    print(key)
    print(mi_diccionario[key])
    print("="*10)
```

```python
for key, value in mi_diccionario.items():
    print(key)
    print(value)
    print("="*10)
```

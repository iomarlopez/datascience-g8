```python
# Ejercicio 1
tupla = (15, 8, 42, 3, 27, 19, 5)
minimo_numero= min(tupla)
maximo_numero = max(tupla)
diff = maximo_numero - minimo_numero
print(
    f"Maximo numero: {maximo_numero}\nMinimo numero: {minimo_numero}\ndiff: {diff}"
)
```

```python
# Ejercicio 2

# Tupla de datos
datos = (5, 8, 5, 1, 9, 5, 2, 8, 5, 3)

# Solicitar número al usuario
numero = int(input("Ingrese un número: "))

# Contar ocurrencias directamente en la tupla
conteo = datos.count(numero)

# Mostrar resultado
print(f"El número {numero} aparece {conteo} veces en la tupla.")
```

```python
#Ejercicio 3

nombres = ('Ana','Luis','Pedro','María','Luis','Carlos')

# Solicitar nombre al usuario
nombre = input("Ingrese un nombre: ")

# Validar si existe en la tupla
if nombre in nombres:
    posicion = nombres.index(nombre)  # devuelve la primera ocurrencia
    print(f"El nombre {nombre} aparece primero en la posición {posicion}.")
else:
    print(f"El nombre {nombre} no existe en la tupla.")
```

```python
#Ejercicio 4

notas = (18, 15, 20, 11, 14, 17, 19)

# Promedio
promedio = sum(notas) / len(notas)

# Máxima y mínima
maxima = max(notas)
minima = min(notas)

# Cantidad de aprobados (nota >= 13)
aprobados = sum(1 for n in notas if n >= 13)

# Mostrar resultados
print(f"Promedio: {promedio:.2f}")
print(f"Nota máxima: {maxima}")
print(f"Nota mínima: {minima}")
print(f"Cantidad de aprobados: {aprobados}")
```

```python
# Ejercicio 5

ventas = [
    ('Lunes', 120),
    ('Martes', 150),
    ('Miércoles', 95),
    ('Jueves', 180),
    ('Viernes', 210)
]

# Día con mayor venta
dia_max, venta_max = max(ventas, key=lambda x: x[1])

# Total de ventas
total = sum(v for _, v in ventas)

# Promedio de ventas
promedio = total / len(ventas)

# Mostrar resultados
print(f"Día con mayor venta: {dia_max} ({venta_max})")
print(f"Total de ventas: {total}")
print(f"Promedio de ventas: {promedio:.2f}")
```

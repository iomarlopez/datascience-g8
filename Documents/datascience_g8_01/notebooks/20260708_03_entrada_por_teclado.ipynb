```python
cuenta = input("ingrese su cuenta")
propina = input("ingrese el procentaje de propina que desea dar")
personas = input ("ingrese la cantidad de personas entre las que desea dividir la cuenta")

monto_propina = (float(cuenta) * float(propina) ) / 100

total = (float(monto_propina) + float(cuenta))

monto_por_persona = float(total) / float(personas)

print("Su total a pagar es:", total)
print("Cada persona tiene que pagar", monto_por_persona)
```

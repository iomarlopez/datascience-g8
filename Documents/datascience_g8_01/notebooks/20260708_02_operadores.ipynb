### Operaores arimeticos

- Suma (+)
- Resta (-)
- Multiplicacion (*)
- Division decimal (/)
- Division entera (//)
- Potencia (**)
- Modulo o resto (%)


```python
numero_1 = 30
numero_2 = 2
print(type(numero_1), type(numero_2))

print("la suma es:", numero_1 + numero_2)
print("la resra es:", numero_1 - numero_2)
print("la multiplicacion es:", numero_1 * numero_2)
print("la division decimal es:", numero_1 / numero_2)
print("la division entera es:", numero_1 // numero_2)
print("la potencia es:", numero_1 ** numero_2)
print("el modulo o resto es:", numero_1 % numero_2)
```

### Operadores Logicos

### Siempre devuelven True o False

- Igual o comparacion (==)
- Diferente (!=)
- Mayor (>)
- Menor (<)
- Mayor o igual (>=)
- Menor o igual (<=)

Operadores para contatenar

- y (and)
- o (or)
- no (not)

```python
numero_1 = 30
numero_2 = 2
numero_3 = 30

print(numero_1 == numero_2)
print(numero_1 == numero_3)
```

```python
print(numero_1 != numero_2)
print(numero_1 > numero_2)
print(numero_1 < numero_3)
```

```python
print(True and True)
```

```python
print(True and False)
```

```python
print(True or False)
```

```python
print(False or False)
```

```python
print(not False)
```

```python
print(not True)
```

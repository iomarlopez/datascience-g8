### Condicionales
SI se cumple esta CONDICION: 
    - Proceso 1
SI se cumple esta otra CONDICION: 
    - Proceso 2
Si no se cumple ninguna de las otras 2 condiciones: 
    - Procesos por defecto o por default




```python
edad = 20

if edad >= 18:
    print("soy un adulto")
elif edad >= 13:
    print("soy un adolescente")
else:
 print("soy un infante")
```

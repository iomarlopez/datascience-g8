###LISTA

- append -> inserta un elemento al final de la lista
- insert -> inserta un elemento con una posicion dada
- remove -> remueve un elemento de la lista
- pop -> remueve el ultimo elemento de la lista
- sort -> ordena la lista
- len -> longitud o contar elementos tiene la lista

```python
list(range(6))
```

```python
lista_numeros = [0, 1, 2, 3, 4, 5]
lista_nombres = ["Juan", "Ana", "Luis"]
lista_mixta = [0, 0.1, 0.5, "Marcos", "Maria"]
```

```python
lista_numeros.append(8)
print(lista_numeros)
```

```python
lista_numeros.append([1,3,5])
print(lista_numeros)
```

```python
lista_numeros.insert(0, "Juan")
print(lista_numeros)
```

```python
## para insert -1 es la penultima posicion
lista_numeros.insert(-1, "Luis")

print(lista_numeros)
```

```python
lista_numeros.remove(4)
print(lista_numeros)
```

```python
lista_numeros.pop()
```
